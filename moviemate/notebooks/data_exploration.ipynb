```python
import sys
import os
import pandas as pd
import matplotlib.pyplot as plt

sys.path.append(os.path.abspath('..'))

from app import load_dataset, partition_data
```

```python
file_path = '../storage/u.data'  
ratings_df = load_dataset(file_path, separation_type='tab')

ratings_df.columns = ['user_id', 'item_id', 'rating', 'timestamp']

train_df, test_df = partition_data(ratings_df, partition_type='stratified')
```

```python
ratings_df.head()
```

### Visualize rating distributions

```python
plt.figure(figsize=(8, 5))
ratings_df['rating'].hist(bins=5)
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Visualize timestamp distribution

```python
plt.figure(figsize=(10, 6))
ratings_df['timestamp'] = pd.to_datetime(ratings_df['timestamp'], unit='s')
ratings_df['timestamp'].hist(bins=50)
plt.title('Timestamp Distribution')
plt.xlabel('Date')
plt.ylabel('Number of Ratings')
plt.grid(True)
plt.show()
```
